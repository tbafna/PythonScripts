```python
%matplotlib inline
import os
import re
import fnmatch
from pathlib import Path
```

```python
def DetectFolder(path):
    
    # if in the correct folder, it gives flag value of 1
    
    if path[-2:] == '\\1' or path[-2:] == '\\2' or path[-2:] == '\\3' or path[-2:] == '\\4' or path[-2:] == '\\5' or path[-2:] == 'MS':
        flag = 1
        print('True')
    else:
        flag = 0
        
    return flag
```

```python
dataPath = r'E:\Data_backup\Data'

flagFolder = 0

for root, dirs, files in os.walk(dataPath):
    
    #print('%s, %s, %s' %(root, dirs, files))
    flagFolder = DetectFolder(root)
    print("")
    
    
    
    if flagFolder > 0:
        for dir_ind, dir_current in enumerate(dirs):
            dir_current_parts = dir_current.split('-')
            print(dir_current_parts)
            
            # if the individual elements have only 1 element, add 0 in front
            for datetime_part_ind, datetime_part in enumerate(dir_current_parts[1:], start=1):
                if len(datetime_part) < 2:
                    dir_current_parts[datetime_part_ind] = '0' + datetime_part
            print(dir_current_parts)
            
            dir_new = '-'.join(dir_current_parts)
            print(dir_new)
            #dirs[dir_ind] = dir_new
            os.rename(os.path.join(root,dir_current), os.path.join(root,dir_new))
        
    
    
```
