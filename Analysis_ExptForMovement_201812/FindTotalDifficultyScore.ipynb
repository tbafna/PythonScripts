```python
%matplotlib
#%matplotlib inline
import os
import csv
import fnmatch
import numpy as np
import re 
import pandas as pd

pd.options.mode.use_inf_as_na = True
```

```python
def ComputeLIXscoreOfPhrase(phrase):
    
    
    B = 1
    sentenceLength = phrase.count(' ')
    words = phrase.split()
    A = sentenceLength
    C = 0
    for word in words:
        if len(word) > 6:
            C = C + 1
        
    LIX = A/B + (C*100)/A
    
    #print(LIX)
    return LIX
```

```python
LIXDifficultList = list()
LIXDifficultScores = list()
LIXEasyList = list()
LIXEasyScores = list() 
LIXMediumList = list()
LIXMediumScores = list()

subjName = r'C:\DTU\Data\201812_ExptToCheckMovementEffect\Data'
j = 0
flagFirstSubj = 0
pupilData = dict()
pupilData['RLCorrelation'] = []

phraseLog_isEmpty = 1

for root, dirs, subfolder in os.walk(subjName):
    if not dirs and 'Test2' in root and 'he' in root:
        print(root)
        if 'tb' in root or 'trial' in root:
            continue
            
        userKeys = None
        gazeLog = None
        keysSelected = None
        
        for file in subfolder:
            if fnmatch.fnmatch(file, 'PhraseLog*'):
                try:
                    
                    fStimPhrase = open(root + '\\' + file, encoding='utf-8')
                    readerStimPhrase = csv.reader(fStimPhrase)
                    stimPhrases = list(readerStimPhrase)
                    stimPhrases.remove(stimPhrases[0])
                    phraseLog_isEmpty = 0
                except:
                    if fStimPhrase is not None:
                        phraseLog_isEmpty = 1
                        fStimPhrase.close()
                    else:
                        print('error in opening the phrase log file')
             # if phraselog not saved - 
            elif phraseLog_isEmpty > 0 and fnmatch.fnmatch(file, 'ReplacementPhraseLog*'):
                
                # find file named ReplacementPhraseLog                
                fStimPhrase = open(root + '\\' + file)
                readerStimPhrase = csv.reader(fStimPhrase)
                stimPhrases = list(readerStimPhrase)
                   
                stimPhrases.remove(stimPhrases[0])
                phraseLog_isEmpty = 0                    
                
            else:
                continue
            
           
            # if all these lists exist
            if not stimPhrases:
                continue
            else:
                
                a = re.compile('(?<=ExptToCheckMovementEffect\\\\Data\\\\)(.*)(?=\\\\2018-1)')
                subjName = a.findall(root)[0]
                print(subjName)
                
                # reduce multiple copies of same phrase to 1 phrase in list
                phraseInd = -1
                # just append another list, if the phrase is a new one
                stimPhrasesNew = list()
                for phraseRow in stimPhrases:
                    phraseInd = phraseInd + 1
                    if phraseInd == 0:
                        continue
                    # 
                    if phraseRow[1] != stimPhrases[phraseInd-1][1]:
                        stimPhrasesNew.append(phraseRow)
                        
                
                # find how many columns used by sentence
                nColumns = 2
                for i in stimPhrasesNew:
                    if len(i)>nColumns:
                        nColumns = max(nColumns,len(i))
                
                
                # give lix score to the sentences                
                for row in stimPhrasesNew:
                    sentence = [','.join(row[1:nColumns])][0]
                    
                    # replace numbers and commas with spaces
                    sentenceForLix = re.sub('[",:;()/\-?!.]', " ", sentence)
                    sentenceForLix = re.sub("[0-9]+.?[0-9]*", " ", sentence)
                    #print(sentenceForLix)
                    
                    # compute LIX score
                    LIX = ComputeLIXscoreOfPhrase(sentenceForLix)
                    
                    # Categorize sentence depending on LIX score
                    if LIX > 60:
                        print('3')
                        LIXDifficultScores.append(LIX)
                        LIXDifficultList.append(sentence)
                    elif LIX < 30:
                        print('1')
                        LIXEasyScores.append(LIX)
                        LIXEasyList.append(sentence)
                    elif LIX > 40 and LIX < 50:
                        print('2')
                        LIXMediumScores.append(LIX)
                        LIXMediumList.append(sentence)
        
```
