```python
import os
import numpy as np
import re
import matplotlib.pyplot as plt
import random
```

```python
# function that computes LIX score 
def ComputeLIXscoreOfPhrase(phrase):
    
    
    B = 1
    sentenceLength = phrase.count(' ')
    words = phrase.split()
    A = sentenceLength
    C = 0
    for word in words:
        if len(word) > 6:
            C = C + 1
        
    LIX = A/B + (C*100)/A
    #print(LIX)
    if LIX < 30:
        print(phrase)
        print('1')
    elif LIX > 55:
        print(phrase)
        print('3')
    
    return LIX
```
