```python
wordList = list()

for sentence in sentences:
    words = sentence.split()
    for word in words:
        
        
        # check if word contains punctuation marks
        if word[-1] == '!' or word[-1] == '.' or word[-1] == ',' or word[-1] == ':' or word[-1] == '?' or word[-1] == '”' or word[-1] == ')' or word[-1] == ';' or word[-1] == '"' or word[-1] == '\'':
            #print(word)
            word = word[:-1]
            #print(word)
            
        if word[0] == '”' or word[0] == '"' or word[0] == '(' or word[0] == '\'':
            #print('nvjkf:    ', word)
            word = word[1:]
        
        # check if word contains number
        if word.isdigit() or word.isdecimal() or '.' in word:
            #print(word)
            continue
        
        
        if word == '!' or word == '?' or word == '.' or word == ',' or word == '-' or word == ':' or word == '&':
            continue
            
        #print(word)
        
        if word == '£50':
            continue
        
        if language == 'english':
            if word == 'I':
                continue
        
        if '(' in word:
            ind = word.index('(')
            word = word[0:ind]
        
        if word[-1] == '!' or word[-1] == '.' or word[-1] == ',' or word[-1] == ':' or word[-1] == '?' or word[-1] == '”' or word[-1] == ')' or word[-1] == ';':
            #print(word)
            word = word[:-1]
            
        if word[0].isdigit():
            continue
            
            
        wordList.append(word)
        
        
        
wordSet = set(wordList)

dictionaryWords = list()
for word in dictionaryWordsRaw:
    dictionaryWords.append(word.rstrip())
```
