```python
from docx import Document
import pandas as pd
import numpy as np
```

```python
document = Document(r"C:\DTU\Data\201911_StudentsFatigue\FlankerTest_Marie\t2.docx")

tables = document.tables

TextDisplayList, CongruencyList, ResponseList, RTList = list(), list(), list(), list()

flagList = 0

for table in tables:
    for row in table.rows:
        for cell in row.cells:
            for paragraph in cell.paragraphs:
                #print(flagList, paragraph.text)
                if flagList == 0:
                    TextDisplayList.append(paragraph.text)
                elif flagList == 1:
                    CongruencyList.append(paragraph.text)
                elif flagList == 2:
                    ResponseList.append(paragraph.text)
                else:
                    RTList.append(float(paragraph.text))
            flagList = flagList + 1
            if flagList == 4:
                flagList = 0
        
```

```python
df = pd.DataFrame(list(zip(TextDisplayList, CongruencyList, ResponseList, RTList)), columns=['TextDisplay', 'Congruency', 'Response', 'ReactionTime'])
```
