```python
%matplotlib inline
import os
import csv
import fnmatch
import numpy as np
import datetime
import re
```

```python
def UserLookedAtScratchPad(keysLookedAt):
    scratchPadKeyTime = list()
    for key in keysLookedAt:
        if 'ScratchPad' in key:
            scratchPadKeyTime.append(key)
    return scratchPadKeyTime
```

```python
# function to convert list of date and time into datetime format list

def timeConversion(timeStrList):
    timeList = list()
    for time in timeStrList:
        time1, t1, t2 = time.partition('+')
        timeList.append(datetime.datetime.strptime(re.sub('[:.T]','-',time1[:-1]), "%Y-%m-%d-%H-%M-%S-%f"))
    return timeList
```

```python
# This function will return the datetime in items which is the closest to the date pivot
def nearestTimePoint(dates, date):
    nearestTP = min([d for d in dates if d<date], key=lambda d: (date - d))
    nearestTPind = dates.index(nearestTP)
    return nearestTP, nearestTPind
```

```python
def findScratchPadText(scratchPadLookedAt, scratchPadList, phraseList):
    phraseScratchedList = list()
    phraseToBeScratchedList = list()
    timeUserLooked = list()
    timeScratchAll = list()
    timeToBeScratchedAll = list()
    
    Veracity = list()
    
    phraseScratchedInd = 0
    phraseToBeScratchedInd = 0
    
    timeStrUserLooked =  [item1[0] for item1 in scratchPadLookedAt]
    timeStrScratch = [item1[0] for item1 in scratchPadList]
    timeStrToBeScracthed = [item2[0] for item2 in phraseList]
    
    timeUserLooked = timeConversion(timeStrUserLooked)
    timeScratchAll = timeConversion(timeStrScratch)
    timeToBeScratchedAll = timeConversion(timeStrToBeScracthed)
    
    for timeWrite in timeUserLooked:
        timeOfPhraseScratchedNow, phraseScratchedInd = nearestTimePoint(timeScratchAll, timeWrite)
        timeOfPhraseToBeScratchedNow, phraseToBeScratchedInd = nearestTimePoint(timeToBeScratchedAll, timeWrite)
        
        phraseScratched = scratchPadList[phraseScratchedInd][1]
        phraseToBeScratched = phraseList[phraseToBeScratchedInd][1]
        
        # are they the same? yes/no -> save in list with the time
        if phraseScratched in phraseToBeScratched:
            Veracity.append([datetime.datetime.strftime(timeWrite, "%Y-%m-%d %H:%M:%S.%f"), str(timeOfPhraseScratchedNow), 'C'])
        else:
            Veracity.append([timeWrite, str(timeOfPhraseScratchedNow), 'INC'])
            
    return Veracity
```

```python
userKeys = None
scratchPad = None
stimPhrase = None

folder = r'C:\DTU\Data\201805_HealthnRehab\TypingData\May16\hc_MS\OptiKeyLogs\2018-5-16-13-42-4'
for file in os.listdir(folder):

    if fnmatch.fnmatchcase(file, 'user_looks_at_key*'):
        try:
            fUserKey = open(folder + '\\' + file, encoding='utf-8')
            readerUserKey = csv.reader(fUserKey)
            userKeys = list(readerUserKey)
            userKeys.remove(userKeys[0])
                    
        except:
            if fUserKey is not None:
                fUserKey.close()
            else:
                print('error in opening the user looks at file')
                        
    elif fnmatch.fnmatch(file, 'ScratchPadLog*'):
        try:                
            fScratchPad = open(folder + '\\' + file, encoding='utf-8')
            readerScratchPad = csv.reader(fScratchPad)
            scratchPad = list(readerScratchPad)
            scratchPad.remove(scratchPad[0])
        except:
            if fScratchPad is not None:
                fScratchPad.close()
            else:
                print('error in opening the scratchpad log file')
                    
    elif fnmatch.fnmatch(file, 'PhraseLog*'):
        try:
            fStimPhrase = open(folder + '\\' + file, encoding='utf-8')
            readerStimPhrase = csv.reader(fStimPhrase)
            stimPhrase = list(readerStimPhrase)
            stimPhrase.remove(stimPhrase[0])
        except:
            if fStimPhrase is not None:
                fStimPhrase.close()
            else:
                print('error in opening the phrase log file')
    else:
        continue
        
    try:
        # if all these lists exist:
        userKeys
        scratchPad
        stimPhrase
        
        # call function to check when scratchpad is looked at and save it in a list
        scratchPadKeyTime = UserLookedAtScratchPad(userKeys)
    
        # for every element, find the time closest and previous to it, and check what was typed 
        # AND what should have been typed
        WrittenTruthList = findScratchPadText(scratchPadKeyTime, scratchPad, stimPhrase)
        WrittenTruthList.insert(0, ['TimeUserLookedAtScratchPad', 'TimeScratchPad', 'Truth'])
        
#         # end for: save list ine excel file
#         with open(folder + '\\MistakeInWriting1.csv', 'w') as csvfile:
#             csvwriter = csv.writer(csvfile)
#             csvwriter.writerows(WrittenTruthList)
        print('done')
        break
    except:
        print('Not yet')
```
